```python

import json

def get_employees_with_country():
    result = []
    company_country_map = {company['name']: company['country'] for company in company_detail_list}
    
    for employee in employee_detail_list:
        company = employee['company']
        country = company_country_map.get(company)
        
        result.append({
            'full_name': employee['full_name'],
            'company': company,
            'country': country
        })
    
    print("ini jawaban no4")
    print(json.dumps(result, indent=4))


get_employees_with_country()
```

```python
import json

def hitung_kebangsaan_karyawan_perusahaan():
    hasil = []
    
    for perusahaan in company_detail_list:
        nama_perusahaan = perusahaan['name']
        kebangsaan_karyawan = {}
        
        for karyawan in employee_detail_list:
            if karyawan['company'] == nama_perusahaan:
                kebangsaan = karyawan['nationality']
                if kebangsaan in kebangsaan_karyawan:
                    kebangsaan_karyawan[kebangsaan] += 1
                else:
                    kebangsaan_karyawan[kebangsaan] = 1
        
        hasil.append({
            'company': nama_perusahaan,
            'employee_nationality': kebangsaan_karyawan
        })
    
    return json.dumps(hasil, indent=4)
print("ini jawaban no 5")
print(hitung_kebangsaan_karyawan_perusahaan())
```
